## Task 10: Sort the data in descending order based on revenue and filter the top 500 movies. Find the measures of central tendancy for the following columns using the filtered data. 1. Budget, 2. Revenue, 3. Runtime. Perform outlier analysis for the above three columns using box plots.

```python
top_500=df.sort_values('revenue',ascending=False).head(500)
top_500
```

```python
def num_data(i):
    mean = top_500[i].mean()
    std = top_500[i].std()
    lower_bound = mean - 2 * std
    upper_bound = mean + 2 * std

    top_500[i] = np.where((top_500[i] < lower_bound) | (top_500[i] > upper_bound), mean, top_500[i])
    print(f'The mean of {i} is : {top_500[i].mean()}')
    print(f'The median of {i} is : {top_500[i].median()}')
    print(f'The mode of {i} is : {top_500[i].mode()[0]}')
    print(f'The standard deviation of {i} is : {top_500[i].std()}')
    print(f'The number of missing values in {i} is : {top_500[i].isnull().sum()}')
    print(f'The skewness of {i} is : {top_500[i].skew()}')
    
```

```python
num_data('budget')
```

```python
num_data('revenue')
```

```python
num_data('runtime')
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.boxplot(x=top_500['revenue'],ax=ax,color='g')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.boxplot(x=top_500['budget'],ax=ax,color='g')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.boxplot(x=top_500['runtime'],ax=ax,color='g')
plt.show()
```
